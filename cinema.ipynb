```python
import pandas as pd
df = pd.read_csv('datafiles/cinema.csv')
df.head(3)
```

```python
#欠損値があるかどうかチェックTrue
df.isnull().any(axis = 0)
```

```python
#欠損値を平均で補充して、df2　に代入
df2 = df.fillna(df.mean())
#穴埋めができたか確認
df2.isnull().any(axis = 0)
```

```python
%matplotlib inline
#SNS2 とsales の散布図の作成
df2.plot(kind = 'scatter',x='SNS2',y='sales')
```
